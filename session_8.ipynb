```python
filepat = r"/home/EDSE-jessewte-workspace/Enron_dataset/maildir/lay-k/all_documents"
```

```python
bydate = {}

for directory, subdirectory, file in os.walk(filepat):
    #print(file)
    for f in file:
        newpat = filepat + "/" + f
        with  open(newpat, "r") as mail:
            data = mail.read()
            out = Parser().parsestr(data)
            dt = out["Date"][5:16]
            if dt in bydate.keys():
                bydate[dt] += 1
            else:
                bydate[dt] = 1
            
         
```
